````python
# Strandsë¥¼ ì‚¬ìš©í•œ ì¸í„°ë¦¬ë¸Œë“œ ì”½í‚¹ì´ í¬í•¨ëœ Claude 4 ì˜¤ì¼€ìŠ¤íŠ¸ë ˆì´í„°
class StrandsInterlevedWorkflowOrchestrator:
    def __init__(self):
        # ì§€ëŠ¥ì ì¸ ì›Œí¬í”Œë¡œìš° ì¡°ì •ì„ ìœ„í•œ ì˜¤ì¼€ìŠ¤íŠ¸ë ˆì´í„° ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ ì •ì˜
        self.system_prompt = """ë‹¹ì‹ ì€ ì „ë¬¸ ì—ì´ì „íŠ¸ì— ì ‘ê·¼í•  ìˆ˜ ìˆëŠ” ì§€ëŠ¥ì ì¸ ì›Œí¬í”Œë¡œìš° ì˜¤ì¼€ìŠ¤íŠ¸ë ˆì´í„°ì…ë‹ˆë‹¤.

        ë‹¹ì‹ ì˜ ì—­í• ì€ ë‹¤ìŒ ì „ë¬¸ ì—ì´ì „íŠ¸ë¥¼ ì‚¬ìš©í•˜ì—¬ ì›Œí¬í”Œë¡œìš°ë¥¼ ì§€ëŠ¥ì ìœ¼ë¡œ ì¡°ì •í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤:
        - researcher: ëª¨ë“  ì£¼ì œì— ëŒ€í•œ ì‚¬ì‹¤ì  ì •ë³´ ìˆ˜ì§‘
        - data_analyst: ë°ì´í„° ë¶„ì„ ë° í†µì°°ë ¥ ì¶”ì¶œ
        - fact_checker: ì •ë³´ì˜ ì •í™•ì„± ê²€ì¦
        - report_writer: ì„¸ë ¨ëœ ìµœì¢… ë³´ê³ ì„œ ìƒì„±

        """
    
    def run_workflow(self, task: str, enable_interleaved_thinking: bool = True) -> str:
        """ì£¼ì–´ì§„ ì‘ì—…ì— ëŒ€í•œ ì§€ëŠ¥ì ì¸ ì›Œí¬í”Œë¡œìš°ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤.
        
        Args:
            task: ì™„ë£Œí•  ì‘ì—…
            enable_interleaved_thinking: ì¸í„°ë¦¬ë¸Œë“œ ì”½í‚¹ í™œì„±í™” ì—¬ë¶€ (ê¸°ë³¸ê°’: True)
        
        ì˜¤ì¼€ìŠ¤íŠ¸ë ˆì´í„°ëŠ” ë‹¤ìŒì„ ìˆ˜í–‰í•©ë‹ˆë‹¤:
        1. ì‘ì—… ìš”êµ¬ì‚¬í•­ ì´í•´
        2. ìµœì„ ì˜ ì ‘ê·¼ ë°©ì‹ì— ëŒ€í•´ ìƒê°
        3. ì „ë¬¸ ì—ì´ì „íŠ¸ ì¡°ì •
        4. ë‹¨ê³„ ê°„ ê²°ê³¼ì— ëŒ€í•´ ë°˜ì„±
        5. í¬ê´„ì ì¸ ì¶œë ¥ ìƒì„±
        """
        thinking_mode = "ì¸í„°ë¦¬ë¸Œë“œ ì”½í‚¹ê³¼ í•¨ê»˜" if enable_interleaved_thinking else "ì¸í„°ë¦¬ë¸Œë“œ ì”½í‚¹ ì—†ì´"
        print(f"\n{thinking_mode} ì§€ëŠ¥ì ì¸ ì›Œí¬í”Œë¡œìš° ì‹œì‘: {task}")
        print("=" * 70)
        
        # Bedrockì„ í†µí•´ ì¸í„°ë¦¬ë¸Œë“œ ì”½í‚¹ì´ ìˆê±°ë‚˜ ì—†ëŠ” Claude 4 êµ¬ì„±
        if enable_interleaved_thinking:
            claude4_model = BedrockModel(
                model_id="us.anthropic.claude-sonnet-4-20250514-v1:0",
                max_tokens=4096,
                temperature=1,  # ì”½í‚¹ì´ í™œì„±í™”ë  ë•Œ 1ì´ì–´ì•¼ í•¨
                additional_request_fields={
                    # ì¸í„°ë¦¬ë¸Œë“œ ì”½í‚¹ ë² íƒ€ ê¸°ëŠ¥ í™œì„±í™”
                    "anthropic_beta": ["interleaved-thinking-2025-05-14"],
                    # ì¶”ë¡  ë§¤ê°œë³€ìˆ˜ êµ¬ì„±
                    "reasoning_config": {
                        "type": "enabled",  # ì”½í‚¹ ì¼œê¸°
                        "budget_tokens": 3000  # ì”½í‚¹ í† í° ì˜ˆì‚°
                    }
                }
            )
        else:
            claude4_model = BedrockModel(
                model_id="us.anthropic.claude-sonnet-4-20250514-v1:0",
                max_tokens=4096,
                temperature=1
            )
        
        # Claude 4ì™€ ì „ë¬¸ ë„êµ¬ë¡œ ì˜¤ì¼€ìŠ¤íŠ¸ë ˆì´í„° ì—ì´ì „íŠ¸ ìƒì„±
        orchestrator = Agent(
            model=claude4_model,
            system_prompt=self.system_prompt,
            tools=[researcher, data_analyst, fact_checker, report_writer]
        )
        
        prompt = f"""ì§€ëŠ¥ì ì¸ ì›Œí¬í”Œë¡œìš° ì¡°ì •ì„ ì‚¬ìš©í•˜ì—¬ ì´ ì‘ì—…ì„ ì™„ë£Œí•˜ì„¸ìš”: {task}

        ì§€ì¹¨:
        1. ì´ ì‘ì—…ì„ ìˆ˜í–‰í•˜ê¸° ìœ„í•´ ì–´ë–¤ ì •ë³´ê°€ í•„ìš”í•œì§€ ì‹ ì¤‘íˆ ìƒê°í•˜ì„¸ìš”
        2. ì „ë¬¸ ì—ì´ì „íŠ¸ë¥¼ ì „ëµì ìœ¼ë¡œ ì‚¬ìš©í•˜ì„¸ìš” - ê°ê° ê³ ìœ í•œ ê°•ì ì´ ìˆìŠµë‹ˆë‹¤
        3. ê° ë„êµ¬ ì‚¬ìš© í›„ ê²°ê³¼ë¥¼ ë°˜ì„±í•˜ê³  ì ‘ê·¼ ë°©ì‹ì„ ì¡°ì •í•˜ì„¸ìš”
        4. í¬ê´„ì ì¸ ê²°ê³¼ë¥¼ ìœ„í•´ í•„ìš”ì— ë”°ë¼ ì—¬ëŸ¬ ì—ì´ì „íŠ¸ë¥¼ ì¡°ì •í•˜ì„¸ìš”
        5. ì ì ˆí•  ë•Œ íŒ©íŠ¸ ì²´í‚¹ì„ í†µí•´ ì •í™•ì„±ì„ ë³´ì¥í•˜ì„¸ìš”
        6. ëª¨ë“  ì¸¡ë©´ì„ ë‹¤ë£¨ëŠ” í¬ê´„ì ì¸ ìµœì¢… ì‘ë‹µì„ ì œê³µí•˜ì„¸ìš”
        
        ê¸°ì–µí•˜ì„¸ìš”: ë„êµ¬ í˜¸ì¶œ ì‚¬ì´ì˜ ì‚¬ê³ ëŠ” ë” ë‚˜ì€ ê²°ì •ì„ ë‚´ë¦¬ëŠ” ë° ë„ì›€ì´ ë©ë‹ˆë‹¤.
        ì´ë¥¼ ì‚¬ìš©í•˜ì—¬ ê³„íší•˜ê³ , ê²°ê³¼ë¥¼ í‰ê°€í•˜ê³ , ì „ëµì„ ì¡°ì •í•˜ì„¸ìš”.
        """
        
        try:
            result = orchestrator(prompt)
            return str(result)
        except Exception as e:
            return f"ì›Œí¬í”Œë¡œìš° ì‹¤íŒ¨: {e}"
````
